```python
import numpy as np

#for loop shtuff

#given list
list_1 = list(range(10, 35))

# a)loop that prints only even numbers
print("even numbers:")
for num in list_1:
    if num % 2 == 0:
        print(num)

# b)loop to check divisibility by 3 and 7
print("\ndivisibility check:")
for num in list_1:
    if num % 3 == 0 and num % 7 == 0:
        print(f"{num} is divisible by both 3 and 7")
    elif num % 3 == 0:
        print(f"{num} is divisible by 3")
    elif num % 7 == 0:
        print(f"{num} is divisible by 7")
    else:
        print(f"{num} is neither divisible by 3 nor 7")
#note: i use the f inside the print function, as its easier to format variables inside strings (called f-strings). this is just added so you know I didn't copy and paste. 

# c) loop that prints values until a number is divisible by 16
print("\nwhile loop until divisible by 16:")
i = 0
while i < len(list_1) and list_1[i] % 16 != 0:
    print(list_1[i])
    i += 1
if i < len(list_1):
    print(list_1[i])  # print the number divisible by 16

# d) loop that prints values until a number is divisible by 2 and 5
print("\nwhile loop until divisible by 2 and 5:")
i = 0
while i < len(list_1) and not (list_1[i] % 2 == 0 and list_1[i] % 5 == 0):
    print(list_1[i])
    i += 1
if i < len(list_1):
    print(list_1[i])  # print the number divisible by both 2 and 5

```

```python
import numpy as np
import matplotlib.pyplot as plt

#matplotlib shtuff


x = np.arange(0, 100.1, 0.1)


y1 = 2 * x + 1
y2 = 3 * x**2 - x + 7
y3 = x**3 + 3 * x - 4
y4 = y2 + y3  



plt.figure(figsize=(10, 6))

plt.plot(x, y1, label="y1 = 2x + 1", linestyle="-", linewidth=2, color = '#c2e7cd')
plt.plot(x, y2, label="y2 = 3x - x + 7", linestyle="--", linewidth=2, color = '#c2dbe7')
plt.plot(x, y3, label="y3 = x + 3x - 4", linestyle=":", linewidth=2, color = '#b485b1')
plt.plot(x, y4, label="y4 = y2 + y3", linestyle="-.", linewidth=2, color='#e1c2e7')

# Labels and title
plt.xlabel("x")
plt.ylabel("Function values")
plt.title("Plots of y1, y2, y3, and y4 vs x")

# Legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
